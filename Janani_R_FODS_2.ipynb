```python
#Bar plot
plt.bar(a['Fat'],a['Sugars'])
plt.xlabel('Fat')
plt.ylabel('Sugars')
plt.title('Fat vs Sugars')
plt.show()
```

```python
plt.plot(a['Caloric Value'],a['Sugars'])
plt.xlabel('Caloric Value')
plt.ylabel('Sugars')
plt.title('Caloric Value vs Sugars')
plt.show()
```

```python
plt.scatter(a['Fat'],a['Sugars'])
plt.xlabel('Fat')
plt.ylabel('Sugars')
plt.title('Fat vs Sugars')
plt.show()
```

```python
sns.boxplot(a['Sugars'])
plt.title('Box plot of Sugars')
plt.show()
```

```python
b=a['Fat']
c=a['Sugars']
plt.hist(b)
plt.xlabel('Fat')
plt.ylabel('Carbohydrate')
plt.title('Histogram')
plt.show()
```
